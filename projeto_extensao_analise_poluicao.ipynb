```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(open("data/who_ambient_air_quality_database_version_2024_(v6.1).xlsx", "rb"), sheet_name="Update 2024 (V6.1)")
df_resume = df[["iso3", "year", "no2_concentration"]]
# df_complete = df_resume.set_index(df_resume.year)
filter_df = df_resume[df_resume["iso3"].isin(["BRA", "USA", "CHN", "CHE", "UZB"])]
grouped_df = filter_df.groupby(["iso3", "year"], as_index=False).sum()
grouped_df
```

```python
grouped_df.set_index("year", inplace=True)
grouped_df.groupby("iso3")["no2_concentration"].plot(legend=True, figsize=(15,10), title="Poluio do Ar - Global", xlabel="Ano", ylabel="Concentrao de Dioxido de Nitrogenio por mcg/m3")
```
